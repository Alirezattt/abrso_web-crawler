```python
# getting all posible shops
common = ["خودرو", "ماشین"]
separate = ["تزئینات","ردیاب","اسپرت","صوتی","امنیت","دزدگیر","لوازم%20اسپرت","صوتی%20و%20تصویری"]
combinations = ["تزئینات","ردیاب","دزدگیر","لوازم%20اسپرت","صوتی%20و%20تصویری","gps","صوتی"]

for i in common :
    for j in separate :
        combinations.append(f"{j}%20{i}")

path = "./torob"
if not os.path.exists(path):
    os.mkdir(path)

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

# Set up Chrome options
options = Options()
# options.add_argument("--headless")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)


all_responses = []
all_ids = set()
for i in combinations :
    Keyword = i.replace("%20", " ")
    url = f"https://api.torob.com/v4/internet-shop/list/?page=0&q={i}&size=1000&show_blocks=true&shop_type=all&source=next_desktop"


    


    while True:
        # Navigate to the API data
        driver.get(url)
        element = driver.find_element(By.XPATH,"/html/body/pre")
        print("getting data from  Keyword :", i)
        data = json.loads(element.text)
        
        for shop in data.get("results", []):
            shop_responses={
                "keyword" : i,
                "id": shop.get("id"),
                "name": shop.get("name"),
                "domain": shop.get("domain"),
                "shop_type": shop.get("shop_type"),
                "city": shop.get("city"),
                "generated_url" : f"https://api.torob.com/v4/internet-shop/details/?id={shop.get('id')}&source=next_desktop"
                }
            all_responses.append(shop_responses)
             
             
        time.sleep(random.uniform(2, 5))

        if data["next"] == None :
            print(f"No next page for {i}")
            print(f"Results : {len(data["results"])}")
            break


        url = data["next"]


with open(f"{path}/all_shops.json", "w", encoding="utf-8") as f:
    json.dump(all_responses, f, ensure_ascii=False, indent=4)
driver.quit()
print("Finished!")
"https://api.torob.com/v4/internet-shop/details/?id=272310&source=next_desktop"
```
